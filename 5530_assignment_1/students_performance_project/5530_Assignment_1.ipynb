```python
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')

import numpy as np
from matplotlib import pyplot as plt

##histogram showing students' performance in math
x_math = df['math score']

plt.hist(x_math, bins=55)

##labels and format
plt.title("Frequency of Math Scores")
plt.xlabel("Math Scores")
plt.ylabel("Students")
plt.show()

##histogram showing students' performance in reading
x_read = df['reading score']

plt.hist(x_read, bins=55)

##labels and format
plt.title("Frequency of Reading Scores")
plt.xlabel("Reading Scores")
plt.ylabel("Students")
plt.show()

##histogram showing students' performance in writing
x_writing = df['writing score']

plt.hist(x_writing, bins=55)

##labels and format
plt.title("Frequency of Writing Scores")
plt.xlabel("Writing Scores")
plt.ylabel("Students")
plt.show()
```

```python
##Pie chart comparing number of students with free/reduced lunch vs standard lunch
reduced = df.loc[df['lunch']== "free/reduced"].count()[0]
standard = df.loc[df['lunch']== "standard"].count()[0]

labels = ['reduced','standard']
colors = ['#00755E','#00C29B']
plt.pie([reduced,standard], labels = labels, colors = colors, autopct = '%.1f')

plt.title("Student Lunches: Free/Reduced vs Standard")

plt.show()
```

```python
##Bar chart of parental level of education and avg test scores of the three subjects (math,reading,writing)

##find mean scores for each test subject
math_avg = df.groupby('parental level of education')['math score'].mean()
reading_avg = df.groupby('parental level of education')['reading score'].mean()
writing_avg = df.groupby('parental level of education')['writing score'].mean()

categories = 6
r = np.arange(categories)
width = 0.28

##create bar plots
plt.bar(r, math_avg, color = 'slateblue',
        width = width, edgecolor = 'black',
        label='Math')
plt.bar(r + width, reading_avg, color = 'darkslateblue',
        width = width, edgecolor = 'black',
        label='Reading')
plt.bar(r + width*2, writing_avg, color = 'royalblue',
        width = width, edgecolor = 'black',
        label='Writing')

##format
plt.xticks([i + width for i in r], math_avg.index, rotation = 'vertical')

plt.ylim(55,80)

##Set title
plt.title('Average Scores by Parental Level of Education')
##Set the x-axis labels
plt.xlabel('Parental Level of Education')

##Set the y-axis labels
plt.ylabel('Average Score')

plt.legend()

plt.show()
```

```python
##Scatter Plot to see correlation between math and reading scores, and math and writing scores - using colors to show gender

male_math = df[df['gender'] == 'male']['math score']
female_math = df[df['gender'] == 'female']['math score']

male_reading = df[df['gender'] == 'male']['reading score']
female_reading = df[df['gender'] == 'female']['reading score']

male_writing = df[df['gender'] == 'male']['writing score']
female_writing = df[df['gender'] == 'female']['writing score']

##Scatter plot comparing female and male math and reading scores
plt.scatter(male_math,male_reading, color = 'deepskyblue', label = 'Male')
plt.scatter(female_math,female_reading, color = 'magenta', label = 'Female')

plt.title('Correlation of Math and Reading scores amongst male and female students')

##Set the x-axis labels
plt.xlabel('Math Scores')

##Set the y-axis labels
plt.ylabel('Reading Scores')

plt.legend()

plt.show()

##Scatter plot comparing female and male math and writing scores
plt.scatter(male_math,male_writing, color = 'deepskyblue', label = 'Male')
plt.scatter(female_math,female_writing, color = 'magenta', label = 'Female')

##Set title
plt.title('Correlation of Math and Writing scores amongst male and female students')

##Set the x-axis labels
plt.xlabel('Math Scores')

##Set the y-axis labels
plt.ylabel('Writing Scores')

plt.legend()

plt.show()
```
